```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## [Daily Treasury Par Yield Curve Rates](https://home.treasury.gov/policy-issues/financing-the-government/interest-rate-statistics)

```python
data = pd.read_csv("alldaily_treasury_rates.csv")
data.Date = pd.to_datetime(data.Date)
data = data.set_index("Date")
data.head()
```

```python
nonan_data = data[data.isnull().sum(axis=1).values==0]
```

```python
start_day = data.index.min()#pd.Timestamp(year=data.index.min().year-1,month=7,day=1)
end_day = data.index.max()#pd.Timestamp(year=data.index.max().year,month=7,day=1)
```

```python
data.plot(colormap="brg", figsize=(20,10), xlim=(start_day, end_day))
```

## 10-2 Year Treasury Yield Spread

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(data['10 Yr']-data['2 Yr'])
#ax.hlines(y=0, xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1], color='r', linestyle='-')
ax.hlines(y=0, xmin=start_day, xmax=end_day, color='r', linestyle='-')
ax.set_xlim(start_day, end_day)
```
